## Gaussian (Normal) Distribution

```python
import numpy as np
from matplotlib import pyplot as plt
```

### 1. Generate and plot normal dist

```python
from scipy.stats import norm

# x-axis for the plot
x_axis = np.arange(-3, 3, 0.001)

# y-axis as the gaussian
y_axis = norm.pdf(x_axis, loc = 0, scale = 1) # y = (x - loc) / scale

# plot data
plt.plot(x_axis, y_axis)
plt.show()
```

### 2. Create a sample of normal distrubited values using 'randn' numpy function

```python
# seed the random number generator
np.random.seed(1)

# generate univariate observations
data = 5 * np.random.randn(10000) + 50

# histogram of generated data
plt.hist(data, bins = 100)
plt.show()
```

```python
# mean value
result = np.mean(data)
print(f'Mean: {result:.3f}')
```

```python
# median
result = np.median(data)
print(f'Median: {result:.3f}')
```

### 3. Variance

```python
plt.plot(x_axis, norm.pdf(x_axis, 0, 0.5), label='0.5')
plt.plot(x_axis, norm.pdf(x_axis, 0, 1), label='1')
plt.legend()
plt.show()
```

```python
# variance
result = np.var(data)
print(f'Variance: {result:.3f}')
```

```python
# standad deviation
result = np.std(data)
print(f'Standard deviation: {result:.3f}')
```
